Consider the wave equation given by:
\begin{equation}
u_{tt}=c^2u_{kk},
\end{equation}
with initial conditions given by $u(x,0)=e^{-x^2}$ and $u_t(x,0)=0$ whose solution is given by:
\begin{equation}
u(x,t) = \frac{1}{2}\left[e^{-(x-ct)^2} + e^{-(x+ct)^2}\right].
\end{equation}

```python
def u_f(x, t, c=1):
    return (1/2) * (np.exp(-(x-c*t)**2) + np.exp(-(x+c*t)**2))
```

```python
x = np.linspace(-10, 10, 1000)
t = np.array([0, 0.25, 0.5, 1, 2.5])
```

```python
plt.figure(figsize=(10, 6))
for t in t:
    plt.plot(x, u_f(x, t), label=f'$t={t}$')
plt.xlabel(r'$x$')
plt.ylabel(r'$u(x, t)$')
plt.legend()
plt.show()
```

Now consider the initial conditions given by:
\begin{equation}
u(x,0) = \begin{cases} 1, & x\geq 0 \\ 0, & x<0 \end{cases}
\end{equation}
and $u_t(x,0)=0$.

```python
def u_f(x, t, c=1):
    u = np.zeros(len(x))
    for i in range(len(x)):
        if x[i] >= c*t:
            u[i] = 1
        elif x[i] <= -c*t:
            u[i] = 0
        elif -c*t < x[i] < c*t:
            u[i] = 1/2
    return u
```

```python
x = np.linspace(-10, 10, 1000)
t = np.array([0, 0.25, 0.5, 1, 5])
```

```python
plt.figure(figsize=(10, 6))
for t in t:
    plt.plot(x, u_f(x, t), label=f'$t={t}$')
plt.xlabel(r'$x$')
plt.ylabel(r'$u(x, t)$')
plt.legend()
plt.show()
```
