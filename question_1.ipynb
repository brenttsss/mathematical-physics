Consider the wave equation given by:
\begin{equation}
u_{tt}=c^2u_{kk},
\end{equation}
with initial conditions given by $u(x,0)=e^{-x^2}$ and $u_t(x,0)=0$ whose solution is given by:
\begin{equation}
u(x,t) = \frac{1}{2}\left[e^{-(x-ct)^2} + e^{-(x+ct)^2}\right].
\end{equation}

```python
def u_f(x, t, c=1):
    return (1/2) * (np.exp(-(x-c*t)**2) + np.exp(-(x+c*t)**2))
```

```python
x = np.linspace(-10, 10, 1000)
t = [0, 0.25, 0.5, 1, 2.5]

plt.figure(figsize=(10, 6))
for t in t:
    plt.plot(x, u_f(x, t), label=f'$t={t}$')
plt.xlabel(r'$x$')
plt.ylabel(r'$u(x, t)$')
plt.legend()
plt.show()
```
